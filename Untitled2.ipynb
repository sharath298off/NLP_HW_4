```python
import torch
import torch.nn as nn
import math
import matplotlib.pyplot as plt

# -----------------------------
# 1. Sample Sentences
# -----------------------------
sentences = [
    "i love nlp",
    "transformers are powerful",
    "attention is important",
    "deep learning is fun",
    "nlp is interesting"
]

# -----------------------------
# 2. Tokenization
# -----------------------------
words = list(set(" ".join(sentences).split()))
word2idx = {w:i+1 for i,w in enumerate(words)}  # start from 1 (0 = PAD)
idx2word = {i:w for w,i in word2idx.items()}

def tokenize(sentence):
    return [word2idx[w] for w in sentence.split()]

tokenized = [tokenize(s) for s in sentences]

# padding
max_len = max(len(s) for s in tokenized)
for i in range(len(tokenized)):
    tokenized[i] += [0] * (max_len - len(tokenized[i]))

inputs = torch.tensor(tokenized)

# -----------------------------
# 3. Embedding
# -----------------------------
d_model = 32
embedding = nn.Embedding(len(words) + 1, d_model)  # +1 for PAD

# -----------------------------
# 4. Positional Encoding
# -----------------------------
def positional_encoding(max_len, d_model):
    pe = torch.zeros(max_len, d_model)
    for pos in range(max_len):
        for i in range(0, d_model, 2):
            pe[pos, i] = math.sin(pos / (10000 ** (i / d_model)))
            if i + 1 < d_model:
                pe[pos, i + 1] = math.cos(pos / (10000 ** (i / d_model)))
    return pe

pe = positional_encoding(max_len, d_model)

# -----------------------------
# 5. Attention Function
# -----------------------------
def attention(Q, K, V):
    scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_model)
    weights = torch.softmax(scores, dim=-1)
    output = torch.matmul(weights, V)
    return output, weights

# -----------------------------
# 6. Multi-Head Attention
# -----------------------------
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, heads=2):
        super().__init__()
        self.heads = heads
        self.d_k = d_model // heads

        self.WQ = nn.Linear(d_model, d_model)
        self.WK = nn.Linear(d_model, d_model)
        self.WV = nn.Linear(d_model, d_model)
        self.fc = nn.Linear(d_model, d_model)

    def forward(self, x):
        Q = self.WQ(x)
        K = self.WK(x)
        V = self.WV(x)

        Q = Q.view(x.size(0), -1, self.heads, self.d_k).transpose(1, 2)
        K = K.view(x.size(0), -1, self.heads, self.d_k).transpose(1, 2)
        V = V.view(x.size(0), -1, self.heads, self.d_k).transpose(1, 2)

        out, attn = attention(Q, K, V)

        out = out.transpose(1, 2).contiguous().view(x.size(0), -1, d_model)
        return self.fc(out), attn

# -----------------------------
# 7. Feed Forward
# -----------------------------
class FeedForward(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, 64),
            nn.ReLU(),
            nn.Linear(64, d_model)
        )

    def forward(self, x):
        return self.net(x)

# -----------------------------
# 8. Encoder Layer
# -----------------------------
class EncoderLayer(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.mha = MultiHeadAttention(d_model)
        self.ff = FeedForward(d_model)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        attn_out, attn = self.mha(x)
        x = self.norm1(x + attn_out)
        ff_out = self.ff(x)
        x = self.norm2(x + ff_out)
        return x, attn

# -----------------------------
# 9. Run Model
# -----------------------------
x = embedding(inputs) + pe.unsqueeze(0)

encoder = EncoderLayer(d_model)
output, attention_weights = encoder(x)

# -----------------------------
# 10. OUTPUTS
# -----------------------------
print("Input Tokens:\n", inputs)

print("\nReadable Tokens:")
for row in inputs:
    print([idx2word.get(i.item(), "<PAD>") for i in row])

print("\nFinal Embeddings:\n", output)

# -----------------------------
# 11. ATTENTION HEATMAP
# -----------------------------
attn = attention_weights[0][0].detach().cpu().numpy()

plt.figure(figsize=(5,4))
plt.imshow(attn)
plt.title("Attention Heatmap")
plt.xlabel("Words")
plt.ylabel("Words")
plt.colorbar()
plt.show()
```
