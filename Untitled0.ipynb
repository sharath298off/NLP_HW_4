```python
import torch
import math

# -----------------------------------
# 1. Scaled Dot-Product Attention
# -----------------------------------
def attention_score(query, key, value):
    depth = query.shape[-1]

    # Step 1: Compute raw attention scores
    score_matrix = torch.matmul(query, key.transpose(-2, -1))
    print("Unscaled Attention Scores:\n", score_matrix)

    # Step 2: Scale scores
    scaled_score_matrix = score_matrix / math.sqrt(depth)
    print("\nScaled Attention Scores:\n", scaled_score_matrix)

    # Step 3: Apply softmax to get probabilities
    attention_distribution = torch.softmax(scaled_score_matrix, dim=-1)
    print("\nAttention Weights:\n", attention_distribution)

    # Step 4: Weighted sum of values
    final_output = torch.matmul(attention_distribution, value)

    return final_output, attention_distribution


# -----------------------------------
# 2. Example Run
# -----------------------------------
torch.manual_seed(0)

query = torch.rand(2, 4)
key = torch.rand(2, 4)
value = torch.rand(2, 4)

result, attention_map = attention_score(query, key, value)

print("\nFinal Output:\n", result)
```
