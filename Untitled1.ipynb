```python
import torch
import torch.nn as nn
import torch.optim as optim
import random
import matplotlib.pyplot as plt

# -------------------- DATA --------------------
text = "hello help hell hello helper hello hell"
chars = sorted(list(set(text)))
char2idx = {ch:i for i,ch in enumerate(chars)}
idx2char = {i:ch for ch,i in char2idx.items()}

vocab_size = len(chars)

# -------------------- HYPERPARAMETERS --------------------
hidden_size = 128
seq_length = 10
lr = 0.003
epochs = 10

# -------------------- DATA PREPARATION --------------------
def create_sequences(text, seq_length):
    inputs, targets = [], []
    for i in range(len(text) - seq_length):
        inputs.append([char2idx[ch] for ch in text[i:i+seq_length]])
        targets.append(char2idx[text[i+seq_length]])
    return inputs, targets

inputs, targets = create_sequences(text, seq_length)

# -------------------- MODEL --------------------
class CharRNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, hidden_size)
        self.rnn = nn.LSTM(hidden_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)

    def forward(self, x):
        x = self.embed(x)
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])
        return out

model = CharRNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=lr)

# -------------------- TRAINING --------------------
loss_history = []

for epoch in range(epochs):
    total_loss = 0
    for x, y in zip(inputs, targets):
        x = torch.tensor(x).unsqueeze(0)
        y = torch.tensor([y])

        output = model(x)
        loss = criterion(output, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(inputs)
    loss_history.append(avg_loss)
    print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}")

# -------------------- LOSS GRAPH --------------------
plt.plot(loss_history)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# -------------------- SAMPLING --------------------
def sample(model, start_str="h", length=200, temperature=1.0):
    model.eval()
    input_seq = torch.tensor([[char2idx[start_str]]])
    result = start_str

    for _ in range(length):
        output = model(input_seq)
        probs = torch.softmax(output / temperature, dim=1).detach().numpy()[0]
        idx = random.choices(range(vocab_size), weights=probs)[0]

        result += idx2char[idx]
        input_seq = torch.tensor([[idx]])

    return result

# -------------------- GENERATION --------------------
print("\nTemp 0.7:\n", sample(model, temperature=0.7))
print("\nTemp 1.0:\n", sample(model, temperature=1.0))
print("\nTemp 1.2:\n", sample(model, temperature=1.2))
```

```python
import torch
import torch.nn as nn
import math
import matplotlib.pyplot as plt

# ---------------- DATA ----------------
sentences = [
    "i love deep learning",
    "transformers are powerful models",
    "neural networks learn patterns",
    "attention improves performance",
    "machine learning is fun",
    "deep learning uses neural nets",
    "nlp tasks use transformers",
    "self attention is important",
    "models learn from data",
    "ai is transforming the world"
]

# Tokenization
all_text = " ".join(sentences).split()
vocab = sorted(set(all_text))
word2idx = {w:i for i,w in enumerate(vocab)}
idx2word = {i:w for w,i in word2idx.items()}
vocab_size = len(vocab)

max_len = max(len(s.split()) for s in sentences)

# Convert sentences to tensor
inputs = []
for s in sentences:
    tokens = s.split()
    idxs = [word2idx[t] for t in tokens]
    idxs += [0] * (max_len - len(idxs))  # padding
    inputs.append(idxs)

inputs = torch.tensor(inputs)

# ---------------- POSITIONAL ENCODING ----------------
def positional_encoding(seq_len, d_model):
    pe = torch.zeros(seq_len, d_model)
    for pos in range(seq_len):
        for i in range(0, d_model, 2):
            pe[pos, i] = math.sin(pos / (10000 ** (i / d_model)))
            if i+1 < d_model:
                pe[pos, i+1] = math.cos(pos / (10000 ** (i / d_model)))
    return pe

# ---------------- MODEL ----------------
class MiniTransformer(nn.Module):
    def __init__(self, vocab_size, d_model=32, heads=4):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, d_model)
        self.pe = positional_encoding(max_len, d_model)

        self.attn = nn.MultiheadAttention(d_model, heads, batch_first=True)

        self.ff = nn.Sequential(
            nn.Linear(d_model, 64),
            nn.ReLU(),
            nn.Linear(64, d_model)
        )

        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        x = self.embed(x)
        x = x + self.pe

        attn_out, attn_weights = self.attn(x, x, x)

        x = self.norm1(x + attn_out)
        x = self.norm2(x + self.ff(x))

        return x, attn_weights

# ---------------- RUN MODEL ----------------
model = MiniTransformer(vocab_size)

out, attn = model(inputs)

print("Input shape:", inputs.shape)
print("Output shape:", out.shape)

print("\nFirst sentence embeddings:\n", out[0])

# ---------------- ATTENTION HEATMAP ----------------
plt.figure(figsize=(6,5))
plt.imshow(attn[0].detach().mean(dim=0), cmap='viridis')
plt.title("Attention Heatmap")
plt.xlabel("Keys")
plt.ylabel("Queries")
plt.colorbar()
plt.show()
```
